```python
from ecephys.data import paths
```

```python
import pandas as pd
```

```python
paths
```

```python
df = pd.DataFrame(columns=['subject', 'condition', 'data', 'path'])
```

```python
df
```

```python
data_type = paths.lfp_bin
data_name = 'lf.bin'
for condition in data_type:
    for subject in data_type[condition]:
        path = data_type[condition][subject]
        df = df.append({'path': path, 'subject': subject, 'condition': condition, 'data': data_name}, ignore_index=True)
```
